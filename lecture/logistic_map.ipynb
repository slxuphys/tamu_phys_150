```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(r,x,n):
    for i in range(n):
        x = r*x*(1-x)
    return x
```

```python
r = np.linspace(0.1,4,2000)
x0 = np.linspace(0.01,0.99,100)
r, x0 = np.meshgrid(r,x0) #generate a grid now r and x are matrices
r = r.ravel() #flatten the matrix to an array
x0 = x0.ravel() #flatten the matrix to an array
plt.plot(r,x0,'.', markersize=0.2)
plt.xlabel('r')
plt.ylabel('x')
fig = plt.gcf()
ax = plt.gca()
line = ax.get_lines()[0]
```

```python
def draw_frame(frame_indx):
    x = logistic_map(r,x0,frame_indx)
    line.set_data(r,x)
```

```python
draw_frame(300)
fig
```

```python
import matplotlib.animation as animation
import IPython

anim = animation.FuncAnimation(fig, draw_frame, frames=100, interval=40)
IPython.display.HTML(anim.to_jshtml())
```

Example 1:

Write a function to implement the map $$x_{n+1} = r \sin(x_n)$$ for number step `n` and rate `r`.

````python
import numpy as np
import matplotlib.pyplot as plt
def sin_map(num_step, x, r):
    x_list=[x]
    for i in range(num_step):
        x=r*np.sin(x)
        x_list.append(x)
    return x_list
````

````python
x_list = sin_map(num_step=100, x=0.1, r=2.8)
plt.plot(x_list)
````

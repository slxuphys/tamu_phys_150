7. The bird's motion follows projectile motion principles. The function simulating the projectile motion is provided below (also in lecture notes). It returns two lists storing the x coordinates and y coordinates of the bird at different time steps. You may need to adjust the maximum time duration of the motion for each chosen combination. (3pt)


```python
def projectile(t_max, num_step, vx, vy):
    pos_x=[0]
    pos_y=[0]
    x=0
    y=0
    dt = t_max/num_step

    for i in range(num_step):
       x = x + vx*dt
       vy = vy -10*dt
       y = y + vy*dt
       pos_x.append(x)
       pos_y.append(y) 
    return pos_x, pos_y
```
